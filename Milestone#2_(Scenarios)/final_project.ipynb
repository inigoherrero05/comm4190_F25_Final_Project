## Pitch

Our idea is to create a chatbot called ***HeyRoommate***: a communication helper for roommates and couples who live together. The app “translates” tense or emotional messages into tailored language that’s better suited to how your specific roommate communicates. For example, if you’re frustrated that `"there have been dishes in the sink for the past week and it's driving me CRAZY"` or want your roommate to `"PLEASE keep it down at night because the noise is annoying"`, *HeyRoommate* can rephrase your message based on tone settings like urgency or seriousness, and it’ll take into account your roommate’s personality preferences such as receptiveness, directness, or sensitivity. The user can also set profiles for each person in the house, so the chatbot can learn how to best get the message across. The goal is to prevent small issues from turning into larger conflicts and help people express themselves neutrally without losing their message.

## Target Users

The target users for ***HeyRoommate*** are roommates and couples who live together in one living space, such as an apartment, condominium, or house.

## Detailed Summary

HeyRoommate is a personalized communication assistant designed to mediate and improve conversations between roommates or partners who share a living space. <br>
The chatbot acts as an emotionally intelligent “translator” that helps users express concerns, requests, or boundaries in ways that align with each recipient’s communication preferences—reducing tension and preventing conflicts from escalating. <br> <br>
**Core Functions:**
* *Message rewriting:* Users input emotionally charged or sensitive messages. HeyRoommate rewrites them into calm, respectful, and effective phrasing based on user-chosen tone settings (e.g. urgency, seriousness, warmth).
* *Profile-based adaptation:* Each roommate has a customizable profile that encodes their communication traits (e.g., direct vs. indirect, sensitive vs. analytical, formal vs. casual). The app learns these over time from feedback from the user, message success, and tone reactions.
* *Contextual awareness:* HeyRoommate references past message outcomes and topics to improve future suggestions—similar to a personalized social memory.
* *Conflict prevention analytics:* The system can flag potentially inflammatory wording or predict misunderstandings, then suggest “smoother” alternatives.
* *Feedback and learning loop:* Users can rate whether the rephrased message “worked” (resolved the issue, received a positive response), allowing the model to refine user profiles continuously.

Both chabot settings and feedback will be inputted into the chatbot via sliders that the user can use to indicate their communication/tone preferences and response effectiveness.

## Data Needed

**From Users:**
* *Message drafts:* Raw user text that needs rewriting.
* *Tone settings:* Sliders for urgency, seriousness, warmth, and directness.
* *Recipient profiles:* Custom or learned attributes (communication style, sensitivity, preferred formality).
* *Feedback:* Ratings on whether the message outcome was positive, neutral, or negative.
* *Interaction history (optional):* Past messages and outcomes to fine-tune recommendations.

**From Web / Other Sources:**
* *Optional pre-trained tone/style data:* Open datasets of conversational tone or politeness models to bootstrap the rephrasing mechanism.
* *Psycholinguistic norms:* Publicly available data on sentiment, empathy, and tone calibration for fine-tuning (e.g., LIWC lexicons).

***HeyRoommate*** will need *analytics logs:* performance data (success rate of rewritten messages, time-to-resolution).
All personal data will be stored locally or encrypted on the server, with an option to auto-delete message histories to protect privacy.

## Role of LLMs

LLMs are the basis of ***HeyRoommate***, crucial for the core functions of this chatbot.

* Language rewriting: Transform emotional or ambiguous drafts into clear, recipient-tailored messages while preserving intent.
* Tone calibration: Interpret tone slider values and apply them in phrasing (e.g., varying levels of directness, empathy, or assertiveness).
* Personality adaptation: Modify wording to match known recipient communication traits (e.g., analytical → factual tone, sensitive → gentle tone).
* Intent extraction: Identify the main topic (e.g., “noise complaint,” “move-out notice”) and conversational goals (“request,” “apology,” “boundary-setting”).
* Dialogue reasoning: Predict potential misinterpretations or conflict triggers before a message is sent, and offer rewording to mitigate them.
* Learning loop support: Summarize user feedback and suggest model adjustments to personalize future outputs. <br>

The LLM operates as a middleware layer—not just a chatbot, but a tone mediator informed by structured user profiles and feedback data.

## What Makes ***HeyRoommate*** unique?

While general LLMs can rewrite text for politeness, they lack memory of relationships, custom tone settings, and feedback-based learning.

| Aspect                | General LLM (e.g., ChatGPT)                        | **HeyRoommate**                                                  |
| --------------------- | -------------------------------------------------- | ---------------------------------------------------------------- |
| **Personalization**   | One-off interactions; no persistent profile memory | Stores profiles for each roommate with communication preferences |
| **Feedback loop**     | No built-in performance learning                   | Tracks message outcomes to adapt phrasing strategies             |
| **Context scope**     | General tone rewriter                              | Specialized for roommate/living-together scenarios               |
| **Structure**         | Conversational chat                                | Guided UI: tone sliders, recipient selection, scenario templates |
| **Conflict modeling** | None                                               | Preemptive tension detection and empathy scoring                 |
| **Privacy design**    | Cloud-only                                         | Option for local processing / anonymized cloud inference         |
