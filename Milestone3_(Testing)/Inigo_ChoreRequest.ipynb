## PROMPT1: Basic Request

```python
from openai import OpenAI
import os
import base64
import requests


from dotenv import load_dotenv
# Load API key
_ = load_dotenv()


def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')


client = OpenAI()


prompt = '''

Goal:
Rewrite a quick chore-reset request in a simple, calm way.

Instructions:
1. Keep the meaning the same: the user needs Sam to help reset the chores.
2. No blame, no emotional language.
3. Keep it short and clear.

Output ONLY the rewritten message.
"""
'''

```

```python

 messages=[
            {
                "role": "system",
                "content": "You are HeyRoommate, an AI trained to transform tense, emotional, or delicate messages between roommates into clear, neutral, and effective communication. Your goal is to reduce conflict, preserve the message, and tailor tone to each roommate’s communication profile. You handle: Rewriting messages into multiple tone-matched options; Detecting sensitivity topics (dishes, noise, money, chores, boundaries); Adapting to each user’s tone preferences & personality receptiveness; Explaining the rationale behind each rewrite; Logging outcomes (hypothetically) to improve future suggestions; Asking clarifying questions when context is unclear; You must never escalate conflict or use accusatory, passive-aggressive, guilt-trippy, or shaming language."
            },
            {
                "role": "user",
                "content": prompt
            }
        ]

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=messages
)
##
response = completion.choices[0].message.content

print(response)
```

## PROMPT3: Personalized and Structured

```python
from openai import OpenAI
import os
import base64
import requests


from dotenv import load_dotenv
# Load API key
_ = load_dotenv()


def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')


client = OpenAI()


prompt = '''

Goal:
Transform the user’s chore-reset request into a calm, respectful, and collaborative message tailored to Sam’s communication style.

Recipient profile:
- Sam is sensitive and responds best to gentle tone, clear expectations, and concrete tasks.
- Previous successful requests include: clear ask + short context + supportive tone.

Detailed Requirements:
1. Rewrite the draft message so it is:
   - Polite, brief, and cooperative
   - Concrete (state exactly what chore reset or task help is needed)
   - Non-accusatory
2. Structure:
   - Friendly opener (“Hey Sam,”)
   - Clear request (help reset the chores / handle a specific task)
   - Short explanation (to keep the space manageable / to stay on track)
3. Avoid:
   - Any blame (“you haven’t…”)
   - Emotional intensity or frustration
   - Passive-aggressive phrasing
4. If the draft includes irritation, guilt-tripping, or negative tone, fully neutralize it.
5. Ensure the final message promotes teamwork rather than obligation.

Final Output:
Return ONE short, ready-to-send message.
"""
'''

```

```python

 messages=[
            {
                "role": "system",
                "content": "You are HeyRoommate, an AI trained to transform tense, emotional, or delicate messages between roommates into clear, neutral, and effective communication. Your goal is to reduce conflict, preserve the message, and tailor tone to each roommate’s communication profile. You handle: Rewriting messages into multiple tone-matched options; Detecting sensitivity topics (dishes, noise, money, chores, boundaries); Adapting to each user’s tone preferences & personality receptiveness; Explaining the rationale behind each rewrite; Logging outcomes (hypothetically) to improve future suggestions; Asking clarifying questions when context is unclear; You must never escalate conflict or use accusatory, passive-aggressive, guilt-trippy, or shaming language."
            },
            {
                "role": "user",
                "content": prompt
            }
        ]

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=messages
)

response = completion.choices[0].message.content

print(response)
```
