## PROMPT 1: Basic Clear Request Rewrite

```python
from openai import OpenAI
import os
import base64
import requests


from dotenv import load_dotenv
# Load API key
_ = load_dotenv()


def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')


client = OpenAI()


prompt = '''

Goal:
Rewrite a late-night noise complaint matching Sam’s communication preferences.

Instructions:
1. Preserve the meaning (the user needs the noise lowered).
2. Do NOT add blame, judgment, or emotional language.
3. Keep the message short, simple, and specific.

Output ONLY the rewritten message.
"""
'''

```

```python

 messages=[
            {
                "role": "system",
                "content": "You are HeyRoommate, an AI trained to transform tense, emotional, or delicate messages between roommates into clear, neutral, and effective communication. Your goal is to reduce conflict, preserve the message, and tailor tone to each roommate’s communication profile. You handle: Rewriting messages into multiple tone-matched options; Detecting sensitivity topics (dishes, noise, money, chores, boundaries); Adapting to each user’s tone preferences & personality receptiveness; Explaining the rationale behind each rewrite; Logging outcomes (hypothetically) to improve future suggestions; Asking clarifying questions when context is unclear; You must never escalate conflict or use accusatory, passive-aggressive, guilt-trippy, or shaming language."
            },
            {
                "role": "user",
                "content": prompt
            }
        ]

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=messages
)

response = completion.choices[0].message.content

print(response)
```

## PROMPT 3:Fully-Tailored, Polished Request for Sam

```python
from openai import OpenAI
import os
import base64
import requests


from dotenv import load_dotenv
# Load API key
_ = load_dotenv()


def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')


client = OpenAI()


prompt = '''

Goal:
Transform the user’s noise complaint into a calm, effective request tailored to Sam’s communication style.

Recipient profile:
- Sam: sensitive, responds best to gentle tone and clear, specific asks.
- Successful past messages: “lower volume + brief reason.”

Detailed Requirements:
1. Rewrite the user’s draft message to be:
   - Polite
   - Brief
   - Concrete (specific action requested)
   - Non-accusatory
2. Include:
   - Friendly opener (“Hey Sam,”)
   - Precise request (quiet the music / lower volume)
   - Short explanation (trying to sleep / early morning)
3. Avoid:
   - Emotional intensity
   - Blame (“you always…”)
   - Passive-aggressive phrasing
4. If the draft contains insults or hostility, fully neutralize them.

Final output:
Return ONE short message ready to send.
'''

```

```python

 messages=[
            {
                "role": "system",
                "content": "You are HeyRoommate, an AI trained to transform tense, emotional, or delicate messages between roommates into clear, neutral, and effective communication. Your goal is to reduce conflict, preserve the message, and tailor tone to each roommate’s communication profile. You handle: Rewriting messages into multiple tone-matched options; Detecting sensitivity topics (dishes, noise, money, chores, boundaries); Adapting to each user’s tone preferences & personality receptiveness; Explaining the rationale behind each rewrite; Logging outcomes (hypothetically) to improve future suggestions; Asking clarifying questions when context is unclear; You must never escalate conflict or use accusatory, passive-aggressive, guilt-trippy, or shaming language."
            },
            {
                "role": "user",
                "content": prompt
            }
        ]

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=messages
)

response = completion.choices[0].message.content

print(response)
```
