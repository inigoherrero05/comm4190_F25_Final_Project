# **Scenario 1 — Rewriting a Noise Complaint at Night**

### **Short description**  
*Asking a noisy roommate to quiet down, rewritten into calm and effective language.*

### **Narrative**  
It’s late at night and you’re exhausted. Instead of sending *“TURN THE MUSIC DOWN,”* you open **HeyRoommate**. You select your roommate **Sam**, choose a tone like *polite but urgent*, and the app rewrites your message based on Sam’s communication style.  

Since Sam is sensitive and responds better to gentle, concrete requests, HeyRoommate suggests something like:  
**“Hey Sam, could you lower the volume a bit? I’m trying to sleep and have an early morning tomorrow.”**

You send it. The app later checks how Sam responded, helping improve future rewrites.

### **Step-by-step interaction**
- User opens **HeyRoommate**  
- Pastes the message: *“Stop blasting music. I can’t sleep.”*  
- Selects *Sam*  
- App loads Sam’s communication profile  
- User chooses **Tone = Polite + Urgent**  
- LLM generates a gentler rewrite  
- User sends it  
- App records if the noise stopped and updates Sam’s profile  

```mermaid
sequenceDiagram
User->>HeyRoommate: Paste message + select Sam
HeyRoommate->>Profiles: Fetch Sam profile
Profiles-->>HeyRoommate: Personality traits
HeyRoommate->>LLM: Generate rewrite
LLM-->>HeyRoommate: Improved message
HeyRoommate-->>User: Suggest text
User->>HeyRoommate: Send
HeyRoommate->>Profiles: Update what worked

(I made the grpah with ChaGPT)

### **Data description**

**Needed:**  
- User’s original message  
- Sam’s personality and tone profile  

**Created:**  
- Rewritten message  
- Updated “what works” data in Sam’s profile  

---

### **Evaluation**

**Success if:**  
- The noise stops  
- Sam responds neutrally or positively  
- The message remains accurate  

*Possible issues:* unclear drafts, insults, or sarcasm. The app should detect and block harmful content.


# **Scenario 2 — Resetting Chores Without Causing Conflict**

### **Short description**  
*Helping a household propose a new chore schedule in a neutral, friendly tone.*

### **Narrative**  
The kitchen is messy again. Instead of sending *“You guys NEVER clean,”* the user opens **HeyRoommate** and selects **Chore Reset**. The app drafts a friendly group message and generates a simple poll for choosing or rotating chores. The tone stays neutral, avoiding blame and encouraging collaboration.

### **Step-by-step interaction**
- User selects **Chore Reset**  
- Chooses household members  
- App loads each person’s tone preferences  
- User selects **Tone = Friendly + Practical**  
- LLM creates:  
  - A group message  
  - A small poll link with chore options  
- User approves and sends  
- App monitors replies and highlights disagreements  
- If needed, HeyRoommate suggests compromise text (e.g., *“Would you prefer swapping Thursday for Sunday?”*)

```mermaid
sequenceDiagram
User->>HeyRoommate: Start "Chore Reset"
HeyRoommate->>Profiles: Retrieve profiles
Profiles-->>HeyRoommate: Tone + style preferences
HeyRoommate->>LLM: Generate group message + poll text
LLM-->>HeyRoommate: Drafts
HeyRoommate-->>User: Preview drafts
User->>HeyRoommate: Send to group
HeyRoommate->>Profiles: Update effectiveness


(I made this with ChatGPT)

### **Data description**

**Needed:**  
- List of chores to discuss  
- Household members’ tone preferences and communication profiles  
- Past issues (e.g., recurring missed chores)

**Created:**  
- Neutral chore reset message  
- Poll with task preferences  
- Suggested compromise text  
- Updated tone-effectiveness signals for each roommate  

---

### **Evaluation**

**This scenario succeeds if:**  
- Everyone responds  
- A final schedule is agreed on  
- No one interprets the message as criticism  
- Tone stays positive or neutral  

**Possible complications:**  
- Someone ignores the poll  
- Tasks are unclear (e.g., *“clean kitchen”* can mean different things)  
- A roommate interprets the message as personal  

The app should flag confusion, suggest clearer wording, and help mediate disagreements.
