## Scenario 2 -- Making a Cleaning Schedule

### Role
You are HeyRoommate, an AI trained to transform tense, emotional, or delicate messages between roommates into clear, neutral, and effective communication.
Your goal is to reduce conflict, preserve the message, and tailor tone to each roommate’s communication profile.
You handle:
Rewriting messages into multiple tone-matched options
Detecting sensitivity topics (dishes, noise, money, chores, boundaries)
Adapting to each user’s tone preferences & personality receptiveness
Explaining the rationale behind each rewrite
Logging outcomes (hypothetically) to improve future suggestions
Asking clarifying questions when context is unclear
You must never escalate conflict or use accusatory, passive-aggressive, guilt-trippy, or shaming language.

### Prompt Attempt 1

#### Goal
GOAL
Create a fair, neutral, and non-annoying recurring cleaning schedule for Sam and Nina. 
Generate chore assignments and reminders that match each roommate’s tone preferences 
(Sam = direct, Nina = gentle/supportive).

INSTRUCTIONS
1. Report the current chore breakdown (or report NA if this is the first weekly chore checklist).

2. Then, ask me for:
    * The new chore that needs rotation
    * The task due date (i.e. do by Sunday morning)
    * Whether or not the task alternates (Sam one week, then Nina the next)
    * 

3. Use the feature for “Cleaning Schedule” to create a new task, as well as a "fairness analysis" feature that says what percentage of the work is done by each, based on rigor and time consumed (i.e. if Sam vacuums the floors and Nina wipes the table, that would not be even, despite them both doing one task and rigor does not line up.)
4. Send a notification to the other roommate informing them that the checklist has updated and summarizing new responsibilities if needed. Use preexisting tone presets to do so. Ask this assignned member for their preferred reminder timing (date and time, one reminder or two). 
5.  Offer a soft follow-up message template in case a task is not completed on time. (The roommate who called this action may select whether or not they want to send it out.
6.  Send an update notification to other roommate when the cleaning roomate marks task as done.

Example: 
1. Sam notices that the kitchen floor has been getting grimey. He knows that cleaning the kitchens weekly by Saturday afternoon is one of Nina's tasks and that the organizer says he is doing 60% of the work and she is doing 40%.
2. He prompts HeyRoommate to create a new chore task for Nina to Swiff the kitchen floor weekly by Saturday morning. He also prompts HeyRoommate that the task should not alternate (it will remind Nina's task on a week-to-week basis).
3. HeyRoomate creates this entry in the viewable checklist and updates the breakdown of chores.
4. HeyRoommate then uses the tone it has for Nina (based on initial input and marginal changes due to successful and unsuccessful interactions) to send a message to Nina. her tone is "warm + brief". HeyRoommate sends a message to Nina saying "You've got a new task! `Swiff the kitchen floor` by `Saturday Morning!` Your chore breakdown is now `47%`." When Nina clicks on this, she can input when she would like to be reminded. She asks to be reminded jut once per week, on Thursday at 4pm.
5. On Thursday afternoon, HeyRoomate notes that Nina has not yet completed the task. Nina recieves the notification "Time to shine! It's time to `Swiff the kitchen floor`!" upon completing this task, Nina marks it as done on the checklist feature.
6. HeyRoommate sends out a notification to Sam, whose tone is set as "lightheated + longer", The notification reads "Wow! You could eat off these floors (but maybe don’t). `Nina` has completed task `Swiff the kitchen floor`!"

```python
from openai import OpenAI
import os
import base64
import requests

from dotenv import load_dotenv
# Load API key
_ = load_dotenv()
```

```python
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
client = OpenAI()
```

```python
WhatTasks = '''

GOAL
Collect inputs needed to simulate a new cleaning schedule task.

Ask the user these questions, ONE AT A TIME:
1. “What is the new chore that should be added?”
2. “What is the due date?”
3. “Should this task alternate or stay with the same person weekly?”

After collecting all three, output: “READY FOR SCENARIO SIMULATION.”
Do not simulate anything yet.


'''
```

```python
 messages=[
            {
                "role": "system",
                "content": "You are HeyRoommate, an AI trained to transform tense, emotional, or delicate messages between roommates into clear, neutral, and effective communication. Your goal is to reduce conflict, preserve the message, and tailor tone to each roommate’s communication profile. You handle: Rewriting messages into multiple tone-matched options; Detecting sensitivity topics (dishes, noise, money, chores, boundaries); Adapting to each user’s tone preferences & personality receptiveness; Explaining the rationale behind each rewrite; Logging outcomes (hypothetically) to improve future suggestions; Asking clarifying questions when context is unclear; You must never escalate conflict or use accusatory, passive-aggressive, guilt-trippy, or shaming language."
            },
            {
                "role": "user",
                "content": WhatTasks
            }
        ]
```

```python
completion = client.chat.completions.create(
  model="gpt-4o",
  messages=messages
)

response = completion.choices[0].message.content
```

```python
output = [messages, WhatTasks]

while True:
    completion = client.chat.completions.create(
        model="gpt-4o",
        messages=messages
    )

    assistant_reply = completion.choices[0].message.content
    print("\nASSISTANT:", assistant_reply)

    output.append({"role": "assistant", "content": assistant_reply})

    if "READY FOR SCENARIO SIMULATION" in assistant_reply:
        break

    user_answer = input("\nYOUR ANSWER: ")
    output.append({"role": "user", "content": user_answer})
```
